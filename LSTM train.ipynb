```python
import pandas as pd
import numpy as np
import calendar
import datetime
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler
from numpy import array
```

```python
def week_of_month(tgtdate):

    days_this_month = calendar.mdays[tgtdate.month]
    for i in range(1, days_this_month):
        d = datetime.datetime(tgtdate.year, tgtdate.month, i)
        if d.day - d.weekday() > 0:
            startdate = d
            break
    return (tgtdate - startdate).days //7 + 1
```

```python
df = pd.read_csv('datasets/train.csv')
df = df.loc[(df['store'] == 1) & (df['item'] == 1)]
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['dayOfWeek'] = df['date'].dt.dayofweek
df['weekOfMonth'] = df['date'].apply(week_of_month)
df.drop(columns=['date'], axis=1, inplace=True)
```

```python
X = df[['year', 'month', 'day', 'dayOfWeek', 'weekOfMonth']]
y = df[['sales']]
```

```python
scalarX, scalarY = MinMaxScaler(), MinMaxScaler()
scalarX.fit(X)
scalarY.fit(y.values.reshape(y.shape[0],1))
X = scalarX.transform(X)
y = scalarY.transform(y.values.reshape(y.shape[0],1))
```

```python
model = Sequential()
model.add(Dense(512, input_dim=5, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='adam')
model.fit(X, y, epochs=1000, verbose=2, shuffle=False, batch_size=400)
```

```python
pred_l, real_l = [], []
for i in range(len(X)):
    predicted = model.predict(array([X[i]]))
    pred = predicted[0][0]
    real = y[i][0]
    pred_l.append(pred)
    real_l.append(real)
    print(f'Predicted: {pred} Actual Value: {real} Diff: {pred-real}')
```

```python
df_res = pd.DataFrame(zip(pred_l, real_l), columns=['pred', 'real'])
df_res['diff'] = (df_res['pred'] - df_res['real'])
df_res.describe().transpose()
```
